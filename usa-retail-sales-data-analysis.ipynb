# USA Retail Sales Data Analysis
## Objective
- Analyze sales trends over the past year.
- Identify regional performance.
- Determine top product categories and specific products driving sales.
- Explore customer segmentation's impact on sales.
- Assess the performance of sales channels.
- Identify the top 10 products by sales volume.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/kaggle/input/sales-dataset-of-usa-updated/sales data.csv'
sales_data = pd.read_csv(file_path)

# Convert date columns to datetime format
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'], format='%d/%m/%Y', errors='coerce')
sales_data = sales_data.dropna(subset=['Order Date'])  # Remove rows with invalid dates


# Extract year, month, and quarter
sales_data['Year'] = sales_data['Order Date'].dt.year
sales_data['Month'] = sales_data['Order Date'].dt.month
sales_data['Quarter'] = sales_data['Order Date'].dt.to_period('Q')
```

## 1. Overall Sales Trends


```python
# Monthly sales trends
monthly_sales = sales_data.groupby('Month')['Sales'].sum()
monthly_sales.plot(kind='bar', color='skyblue', figsize=(10, 6), title='Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()

# Quarterly sales trends
quarterly_sales = sales_data.groupby('Quarter')['Sales'].sum()
quarterly_sales.plot(kind='bar', color='green', figsize=(10, 6), title='Quarterly Sales Trends')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.show()
```

## 2. Sales Performance by Region


```python
# Sales by region
region_sales = sales_data.groupby('Region')['Sales'].sum()
region_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), title='Sales Distribution by Region')
plt.ylabel('')
plt.show()
```

## 3. Contribution of Product Categories


```python
# Sales by product category
category_sales = sales_data.groupby('Category')['Sales'].sum()
category_sales.plot(kind='bar', color='orange', figsize=(10, 6), title='Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

## 4. Customer Segmentation Analysis


```python
# Sales by segment
segment_sales = sales_data.groupby('Segment')['Sales'].sum()
segment_sales.plot(kind='bar', color='purple', figsize=(10, 6), title='Sales by Customer Segment')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.show()
```

```python
## 5. Performance of Sales Channels
```

```python
# Analyzing sales channels
channel_sales = sales_data.groupby('Ship Mode')['Sales'].sum()
channel_sales.plot(kind='bar', color='cyan', figsize=(10, 6), title='Sales by Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Total Sales')
plt.show()
```

## 6. Top 10 Products by Sales Volume


```python
# Top 10 products
top_products = sales_data.groupby('Product Name')['Sales'].sum().nlargest(10)
top_products.plot(kind='bar', color='red', figsize=(10, 6), title='Top 10 Products by Sales Volume')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.show()
```
